# **f. Formula del coeficiente de variación**
La fórmula del **coeficiente de variación (CV)** se expresa de la siguiente manera:

$$
CV = \frac{\sigma}{\mu} \times 100
$$

Donde:
- $\sigma$ es la **desviación estándar** de los datos.
- $\mu $ es la **media** del conjunto de datos.

El coeficiente de variación es una medida relativa de dispersión que expresa la **variabilidad de los datos en relación con la media**. Su importancia radica en que permite **comparar la variabilidad** entre distintos conjuntos de datos, independientemente de sus unidades de medida o magnitud. Esto es útil, por ejemplo, cuando se quiere comparar la consistencia de diferentes inversiones, experimentos, o mediciones, ya que una menor variabilidad puede indicar mayor estabilidad o precisión.
